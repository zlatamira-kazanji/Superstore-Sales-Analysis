```python
import pandas as pd  
import numpy as np
import matplotlib.pyplot as plt 
df = pd.read_csv("train.csv")  

#print(df.head())   
print(df.info()) 
print(df.describe())   
```

```python

plt.hist(np.log1p(df["Sales"]), bins=50, color="skyblue", edgecolor="black")

plt.title("Распределение продаж (логарифмическая шкала)")
plt.xlabel("log(Sales + 1)")  # Ось X: логарифм продаж
plt.ylabel("Количество заказов")  # Ось Y: частота встречаемости

plt.grid(axis="y", alpha=0.3)

plt.show()
```
